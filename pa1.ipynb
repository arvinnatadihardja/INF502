to find number of matches in sequences, i have concat sequences with '-' and the iterate over sequences and see it 
characters are same score increments.
to find macchain, i keep track of prev matching characters and if next match is consequtive with previous it increments chain = chain + 1

```python
def maxChain(seq1,seq2):
    chain = 0
    prev = 0
    for i in range(0,len(seq1)):
        c1 = seq1[i]
        c2 = seq2[i]
        if c1 == c2 and prev == 0:
            prev = i
            chain =  chain + 1
        elif c1 == c2  and i ==  (prev + 1):
            prev = i
            chain  = chain + 1
    return chain

def numberofmatches(seq1,seq2):
    score = 0
    for i in range(0,len(seq1)):
        c1 = seq1[i]
        c2 = seq2[i]
        if c1 == c2:
            score = score+1
    return score
```

```python
def shiftSeq1(s1,s2,shifts):
    print("Before shifting")
    score = numberofmatches(s1,s2)
    print(s1)
    print(s2)
    print("Score:",score)
    print("\n")

    for i in range(0,shifts):
        print("shifting sequence1 by "+str(i+1))
        s1 = '-'+s1
        s2 = s2+'-'
        score = numberofmatches(s1,s2)
        print(s1)
        print(s2)
        print("Score:",score)
        print("\n")

def shiftSeq2(s1,s2,shifts):
    print("Before shifting")
    score = numberofmatches(s1,s2)
    print(s1)
    print(s2)
    print("Score:",score)
    print("\n")

    for i in range(0,shifts):
        print("shifting sequence2 by "+str(i+1))
        s1 = s1+''
        s2 = '-'+s2
        score = numberofmatches(s1,s2)
        print(s1)
        print(s2)
        print("Score:",score)
        print("\n")
        
def chainSeq1(s1,s2,shifts):

    for i in range(0,shifts):
        print("shifting sequence1 by "+str(i+1))
        s1 = '-'+s1
        s2 = s2+'-'
        print(s1)
        print(s2)
        print("Max Chain:",maxChain(s1,s2))
        print("\n")

def chainSeq2(s1,s2,shifts):
   
    for i in range(0,shifts):
        print("shifting sequence2 by "+str(i+1))
        s1 = s1+''
        s2 = '-'+s2
        print(s1)
        print(s2)
        print("Max Chain:",maxChain(s1,s2))
        print("\n")
    
```

```python
def main():
    shifts = input("Enter Max Shifts: ")
    seq1 = input("Enter seq1.txt file: ")
    seq2 = input("Enter seq3.txt file: ")
    approach = input("Enter 1 for MAX chains, 2 for Number of Matches: ")
    
    try:
        f=open(seq1, "r")
        s1 = f.read()
        f.close()
    
        f=open(seq2, "r")
        s2 = f.read()
        f.close()
        
        if approach == "1":
            shiftSeq1(s1,s2,shifts)
            shiftSeq2(s1,s2,shifts)
        else:
            chainSeq1(s1,s2,shifts)
            chainSeq2(s1,s2,shifts)
    except:
        print("unable to read input files")
    

if __name__ == '__main__':
    main()
```
